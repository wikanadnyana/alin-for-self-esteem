```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
"""
Solution for problem number 2e page 12 :
"""

coef = np.linspace(-100, 100, 1000000)

y1 = (2 - math.pi * coef) / -5
y2 = -1 + math.pi * coef

x_solution = 8 / 5 * math.pi
y_solution = -1/4


fig, ax = plt.subplots()
plt.title("Solution Problem Number 2e page 12")
plt.xlabel("x")
plt.ylabel("y")
ax.set_xlim([-100,100])
ax.set_ylim([-100,100])
ax.plot(coef, y1, c='green')
ax.plot(coef, y2, c='brown')
plt.axvline(x=x_solution, color='purple', linestyle='--')
plt.axhline(y=y_solution, color='purple', linestyle='--')
```

```python
"""
Solution for problem number 2f page 12 :
"""

coef = np.linspace(-100, 100, 1000000)

e = math.e

y1 = (2 - e*coef) / -e
y2 = -1 * coef

x_solution = 1 / e
y_solution = -1 / e


fig, ax = plt.subplots()
plt.title("Solution Problem Number 2f page 12")
plt.xlabel("x")
plt.ylabel("y")
ax.set_xlim([-100,100])
ax.set_ylim([-100,100])
ax.plot(coef, y1, c='green')
ax.plot(coef, y2, c='brown')
plt.axvline(x=x_solution, color='purple', linestyle='--')
plt.axhline(y=y_solution, color='purple', linestyle='--')
```
