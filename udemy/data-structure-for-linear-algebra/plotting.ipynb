```python
import numpy as np
import matplotlib.pyplot as plt
```

## First Problem Example

```python
t = np.linspace(0, 40, 1000)
t
```

```python
d_r = 2.5 * t
```

```python
d_s = 3 * (t-5)
```

```python
fig, ax = plt.subplots()
plt.title("A Bank Robber Caught")
plt.xlabel("time in minutes")
plt.ylabel("distance in km")
ax.set_xlim([0,40])
ax.set_ylim([0,100])
ax.plot(t, d_r, c='green')
ax.plot(t, d_s, c='brown')
plt.axvline(x=30, color='purple', linestyle='--')
plt.axhline(y=75, color='purple', linestyle='--')
```

## Second Problem Example

```python
t2 = np.linspace(0, 100, 1000)
```

```python
e_apr = t2
e_may = 4 * (t2 - 30)
```

```python
fig, ax = plt.subplots()
plt.title("Solar Panel Mark Design Problem")
plt.xlabel("time in day")
plt.ylabel("energy")
ax.set_xlim([1,60])
ax.set_ylim([0,100])
ax.plot(t2, e_apr, c='green')
ax.plot(t2, e_may, c='brown')
plt.axvline(x=40, color='purple', linestyle='--')
plt.axhline(y=40, color='purple', linestyle='--')
```
